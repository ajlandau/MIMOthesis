```python

def roomSections(array):
    px = np.size(array, axis = 0)
    py = np.size(array, axis = 1)
    sect = 3
    divx = (px+sect //2)//sect
    divy = (py+sect//2)//sect
    corner = array[4:divx, 4:divy]
    center = array[divx+1:divx*2, divy+1:divy*2]
    yside = array[4:divx, divy+1:divy*2]
    xside = array[divx+1:divx*2, 4:divy]
    return corner, center, yside, xside
corner_ant8c, center_ant8c, yside_ant8c, xside_ant8c = roomSections(10*np.log10(corner8antenna))
corner_counts_c, corner_bins_c = CDF_generation(corner_ant8c)
center_counts_c, center_bins_c = CDF_generation(center_ant8c)
yside_counts_c, yside_bins_c = CDF_generation(yside_ant8c)
xside_counts_c, xside_bins_c = CDF_generation(xside_ant8c)

corner_ant8s, center_ant8s, yside_ant8s, xside_ant8s = roomSections(10*np.log10(side8antenna))
corner_counts_s, corner_bins_s = CDF_generation(corner_ant8s)
center_counts_s, center_bins_s = CDF_generation(center_ant8s)
yside_counts_s, yside_bins_s = CDF_generation(yside_ant8s)
xside_counts_s, xside_bins_s = CDF_generation(xside_ant8s)
```
